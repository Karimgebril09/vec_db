```python
nbits = 3  # number of hyperplanes and binary vals to produce
d = 64  # vector dimensions
```

```python
import numpy as np
np.random.seed(0)
# create a set of 4 hyperplanes, with 2 dimensions
#  -.5 to center our array values around the origin (0, 0).
plane_norms = np.random.rand(nbits, d) - .5
plane_norms
```

```python
a = np.random.rand(d)
b = np.random.rand(d)
c = np.random.rand(d)
dataset=[a,b,c]
```

```python
a_dot = np.dot(a, plane_norms.T)
b_dot = np.dot(b, plane_norms.T)
c_dot = np.dot(c, plane_norms.T)
a_dot
```

```python
a_dot = a_dot > 0
b_dot = b_dot > 0
c_dot = c_dot > 0
c_dot
```

```python
a_dot = a_dot.astype(int)
b_dot = b_dot.astype(int)
c_dot = c_dot.astype(int)
a_dot
```

```python
vectors = [a_dot, b_dot, c_dot]
buckets = {}  # hash table
i = 0

for i in range(len(vectors)):
    # convert from array to string
    hash_str = ''.join(vectors[i].astype(str))
    # create bucket if it doesn't exist
    if hash_str not in buckets.keys():
        buckets[hash_str] = []
    # add vector position to bucket
    buckets[hash_str].append(i)

print(buckets)
```

```python
def HammingDistance(s1: str, s2: str) -> int:
    """ Compute the Hamming distance between two bitstrings """
    assert len(s1) == len(s2)
    dist = 0
    for i in range(len(s1)):
        if s1[i] != s2[i]:
            dist += 1
    return dist
```

```python
def find_closest_bucket(query_hash_str, buckets):
    if query_hash_str in buckets:
        return buckets[query_hash_str]
    else:
        # Find the closest bucket using Hamming distance
        min_distance = float('inf')
        closest_bucket = None
        for bucket_key in buckets.keys():
            distance = HammingDistance(query_hash_str, bucket_key)
            if distance < min_distance:
                min_distance = distance
                closest_bucket = buckets[bucket_key]
        return closest_bucket
```
